```python
#import requiered libraries 
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
insurance = pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv")
insurance
```

#features
Our independent variable are age, sex, bmi, children, smoker, region
#labels 
Our dependent variable is charges 

##before passing the features to our neural network we have to turn string features like **sex**, **smoker**, **region** to numbers

```python
insurance_one_hot = pd.get_dummies(insurance)
insurance_one_hot.head()
```

### Create **X** and **Y** values
### training and test set 
### build network 
###compile it
###fit it
### evaluate it 
###tweak it 

```python
X = insurance_one_hot.drop("charges", axis = 1)
Y = insurance_one_hot["charges"]
```

```python
X.head(), Y.head()
print(len(X))
print(len(Y))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
len(X), len(X_train), len(X_test)
```

```python
# built, compile and fit the model 
tf.random.set_seed(42)
insurance_model = tf.keras.Sequential([
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])
insurance_model.compile(loss= tf.keras.losses.mae,
                        optimizer = tf.keras.optimizers.SGD(),
                        metrics = ['mae'])
insurance_model.fit(X_train, Y_train, epochs = 100)
```

```python
insurance_model.evaluate(X_test, Y_test)
```

#time to improve our model

```python
tf.random.set_seed(42)
insurance_model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])
insurance_model_2.compile(loss= tf.keras.losses.mae,
                        optimizer = tf.keras.optimizers.Adam(),
                        metrics = ['mae'])
insurance_model_2.fit(X_train, Y_train, epochs = 100, verbose = 1)
```

```python
insurance_model_2.evaluate(X_test, Y_test)
```

```python
tf.random.set_seed(42)
insurance_model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])
insurance_model_3.compile(loss= tf.keras.losses.mae,
                        optimizer = tf.keras.optimizers.Adam(),
                        metrics = ['mae'])
history =insurance_model_3.fit(X_train, Y_train, epochs = 200, verbose = 1)
```

```python
insurance_model_3.evaluate(X_test, Y_test)
```

#time to normalize our data for better acuracy

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder 

ct = make_column_transformer(
    (MinMaxScaler(), ["age", "bmi", "children"]),
    (OneHotEncoder(handle_unknown = 'ignore'), ["sex", "smoker", "region"])
)

#create X and Y 
X = insurance.drop("charges", axis =1)
Y= insurance["charges"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)

ct.fit(X_train)
X_train_normal = ct.transform(X_train)
X_test_normal = ct.transform(X_test)
```

```python
X_train.loc[0]
```

```python
X_train_normal[0]
```

```python
tf.random.set_seed(42)

insurance_model_4 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)

])

insurance_model_4.compile (loss = tf.keras.losses.mae,
                           optimizer = tf.keras.optimizers.Adam(),
                           metrics = ["mae"])

insurance_model_4.fit(X_train_normal, Y_train, epochs=100)
```

```python
insurance_model_4.evaluate(X_test_normal, Y_test)
```
