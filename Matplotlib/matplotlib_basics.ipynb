##Matplotlib library
Matplotlib is a popular open-source data visualization library in Python that allows users to create a wide range of high-quality plots, including line plots, scatter plots, bar plots, histograms, and more. It was created by John Hunter and is widely used in scientific and engineering fields for data analysis, exploration, and presentation. Matplotlib provides a flexible and powerful API for customization and fine-tuning of plots to suit users' needs. It is also compatible with other Python libraries and frameworks, such as NumPy, Pandas, and Seaborn, making it a versatile tool for data science and machine learning projects.

##import required libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x_1 = np.linspace(0, 5, 10) #creates a NumPy array with 10 evenly spaced values between 0 and 5, inclusive.
y_1 = x_1**2 #The code y_1 = x_1**2 then takes each value in x_1, squares it, and assigns the resulting value to the corresponding index in y_1.
```

##creating a plot 

```python
#The code plt.plot(x_1, y_1) will plot a parabolic curve using Matplotlib, 
#with the values in x_1 on the x-axis and the corresponding squared values in y_1 on the y-axis. 
#The plt.title('myplot') code will set the title of the plot to "myplot", while plt.xlabel('time') 
#and plt.ylabel('velocity') will label the x-axis and y-axis with the text "time" and "velocity" respectively.
plt.plot(x_1,y_1)
plt.title('myplot')
plt.xlabel('time')
plt.ylabel('velocity')
```

##To print multiple plots

```python
#The below code will produce a figure with four subplots arranged in a 2x2 grid. 
#The first subplot (plt.subplot(2,2,1)) will plot the parabolic curve in red (plt.plot(x_1, y_1, 'r')), 
#the second subplot (plt.subplot(2,2,2)) will plot the same curve in blue (plt.plot(x_1, y_1, 'b')), 
#and the third subplot (plt.subplot(2,2,3)) will plot the curve in red again (plt.plot(x_1, y_1, 'r')).
plt.subplot(2,2,1)

plt.plot(x_1, y_1, 'r')
plt.subplot(2,2,2)
plt.plot(x_1, y_1, 'b')
plt.subplot(2,2,3)
plt.plot(x_1, y_1, 'r')
```

##multiple plots in one plot with add_axes function

```python
#The first subplot (axes_1) plots the x_1 array against the y_1 array with a label of "x/x2" and also plots the y_1 array against x_1 with a label of "y". 
#The x-axis is labeled "dats" and the y-axis is labeled "y". 
#The legend is positioned using loc=0, which means that Matplotlib will try to place it in the "best" location.
# The second subplot (axes_2) plots the same data as the first subplot, but with a subplot size of 0.2 by 0.2 and positioned at the center-right of the figure using the add_axes() method.
fig_1 = plt.figure(figsize=(5,4), dpi=100)
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9])
axes_1.set_xlabel('dats')
axes_1.set_ylabel('y')
axes_1.plot(x_1, y_1, label='x/x2')
axes_1.plot(y_1, x_1, label = 'y')
axes_1.legend(loc=0)
axes_2 = fig_1.add_axes([0.5,0.5,0.2,0.2])
axes_2.plot(x_1, y_1, label = 'x')
```

##subplots

```python
#The code creates a figure with three subplots arranged in a single row (nrows=1, ncols=3) with a total figure size of 9x4 inches (figsize=(9,4)).
# The axes_3 variable is a NumPy array that contains the three subplot objects.

#The axes_3[1] command selects the subplot in the middle of the row (since indexing starts from 0), 
#and the axes_3[1].title.set_text('plot 2') command sets the title of this subplot to "plot 2". 
#The axes_3[1].set_xlabel('x') and axes_3[1].set_ylabel('y') commands label the x-axis and y-axis of the subplot, respectively.
# The axes_3[1].plot(x_1, y_1) command plots the x_1 array against the y_1 array on the selected subplot. The axes_3[1].set_facecolor('pink') command sets the face color of the subplot to pink.

#The plt.tight_layout() command adjusts the spacing between the subplots to optimize the use of space in the figure.
fig_2 , axes_3 = plt.subplots(figsize=(9,4), nrows = 1, ncols=3)
plt.tight_layout()
axes_3[1].title.set_text('plot 2')
axes_3[1].set_xlabel('x')
axes_3[1].set_ylabel('y')
axes_3[1].plot(x_1, y_1)
axes_3[1].set_facecolor('pink')
```

```python
#The following code creates a figure with a single subplot (axes_4) and sets its dimensions to be 8x4 inches (figsize=(8,4)). 
#The axes_4.plot() command plots the x_1 array against the y_1 array with a navy color (color='navy'), a transparency of 1 (alpha=1),
# a line width of 2 (lw=2), a dashed line style (ls='-.'), and circular markers (marker='o') with a size of 7 (markersize=7). 
#The markers are filled with yellow (markerfacecolor='y') and have a red edge color (markeredgecolor='r') and an edge width of 1 (markeredgewidth='1').

#The axes_4.set_facecolor('pink') command sets the face color of the subplot to pink.

#The axes_4.set_xlim([0,3]) command sets the x-axis limits of the subplot to be between 0 and 3, 
#effectively zooming in on the left-hand side of the plot. The axes_4.set_xlim([0,8]) command sets the y-axis limits of the subplot to be between 0 and 8, 
#effectively zooming in on the upper part of the plot.

#The axes_4.grid(True, color='0.6', dashes=(5,2,1,2)) command adds a grid to the subplot with a light gray color (color='0.6') and a dash pattern of 5 points on, 2 points off, 1 point on, 2 points off.

#The axes_4.text() commands add two lines of text to the subplot.
# The first line uses LaTeX markup to display a mathematical equation with Greek letters (r'$\alpha \beta \sigma \omega \epsilon \pi \mu \theta \lambda$'). 
#The second line uses LaTeX markup to display another equation with Greek letters and summation notation (r'$\delta_i \gamma^{i} \sum_{i=0} ^\infty x_i \frac{3}{4}$').
fig_3 = plt.figure(figsize=(8,4))
axes_4 = fig_3.add_axes([0,0,1,1])
axes_4.plot(x_1, y_1, color = 'navy', alpha = 1, lw = 2, ls='-.', marker = 'o',markersize = 7, markerfacecolor = 'y', markeredgecolor = 'r', markeredgewidth= '1' )
axes_4.set_facecolor('pink')

#to zoom in to a plot
axes_4.set_xlim([0,3])
axes_4.set_xlim([0,8])

axes_4.grid(True, color = '0.6', dashes=(5,2,1,2))

axes_4.text(0,23,
            r'$\alpha \beta \sigma \omega \epsilon \pi \mu \theta \lambda$')

axes_4.text(0,18,
            r'$\delta_i \gamma^{i} \sum_{i=0} ^\infty x_i \frac{3}{4}$') # alot more functions with tex markup..
```

#pie charts

```python
import random

fig_6 = plt.figure(figsize = (8,4), dpi = 100)
axes_6 = fig_6.add_axes([0.1, 0.1, 0.9,0.9])

types = ['wow','something','nothing','wow','something','nothing','nothing','water','summer','winter','spring','nothing','nothing']

num = [133,56,78,44,33,22,22,45,34,67,34,12,34]

color = []
for i in range(13):
  rgb = (random.uniform(0, .5), random.uniform(0, .5), random.uniform(0, .5))
  color.append(rgb)

explode = [0] * 13
explode[0] = 0.2

wedges, texts, autotexts = plt.pie(num, explode = explode, labels = types, colors = color, autopct = '%1.0f%%', shadow = True, startangle = 150, textprops = dict(color='w'))
plt.legend(wedges, types, loc = 'right', bbox_to_anchor=(1,0,0.5,1))


#The code import random, fig_6 = plt.figure(figsize=(8,4), dpi=100), 
#axes_6 = fig_6.add_axes([0.1, 0.1, 0.9,0.9]), 
#types = ['wow','something','nothing','wow','something','nothing','nothing','water','summer','winter','spring','nothing','nothing'],
# num = [133,56,78,44,33,22,22,45,34,67,34,12,34] creates two lists types and num with the labels and sizes of the wedges in the pie chart.
# The random.uniform(0, .5) function generates random numbers between 0 and 0.5 which are used to create a list of colors for each wedge in the pie chart. 
#The explode variable is set to a list of zeros to indicate that none of the wedges are exploded, 
#except for the first wedge which is exploded by setting its value to 0.2. The plt.pie() function creates the actual pie chart,
# with the specified attributes such as the explode parameter to control which wedge(s) is exploded, 
#the labels parameter to specify the labels for each wedge, the colors parameter to specify the color for each wedge, 
#the autopct parameter to display the percentage of each wedge, the shadow parameter to add a shadow effect, 
#the startangle parameter to specify the angle at which the first wedge starts, and the textprops parameter to set text properties. 
#Finally, the plt.legend() function adds a legend to the pie chart with the loc and bbox_to_anchor parameters used to specify the position of the legend.
```
